## Activation Function

```python
import numpy as np
import matplotlib.pyplot as plt
# Linear Activation
def linear(x):
  y=x
  return y

# signum activation
# let theta(threshold = 0)
def signum(x):
  if x==0:
    y=0
  elif x>0:
    y=1
  else:
    y=-1

  return y
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x):
    return np.maximum(0.01*x, x)

def elu(x):
    return np.maximum(1.0*(np.exp(x)-1), x)
```

## Perceptron

```python
def perceptron(weight, input, target):
  n = 4
  for i in range(n):
    w_transpose = np.transpose(weight)
    x = input[i, :]
    print("x = ", x)
    d = target[i]
    y_in = np.dot(w_transpose, x)
    print("y_in = ", y_in)
    print("d = ", d)
    if (d>0 and y_in < 0):
      weight = weight + x

    if (d<=0 and y_in >= 0):
      weight = weight - x
    print("weight = ", weight, "\n")

  return weight
```

# TRAIN

```python
import numpy as np

# Binary OR
X = np.array([[0,0],[0,1],[1,0],[1,1]], dtype=float)
y = np.array([[0],[1],[1],[1]], dtype=float)

W = np.random.rand(2)

for epoch in range(100):
  print("Perptron")
  W=perceptron(W, X, y)
print("Weight= ", W)
```
